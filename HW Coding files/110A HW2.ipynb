We write the code down for initial point $x_{0}=(1.2,1.2)$. Then , we consider the case where $x_{0}=(-1.2,1)$

```python
import numpy as np
import matplotlib.pyplot as plt

f5=lambda x,y: 100*(y-x**2)**2+(1-x)**2

Df5=lambda x,y: np.array([-400*x*(y-x**2)-2*(1-x),200*(y-x**2)])

def WolfeI(alpha,f,x,p,c1=0.1):
    LHS=f(x[0]+alpha*p[0],x[1]+alpha*p[1])
    RHS=f(x[0],x[1])-c1*alpha*np.dot(p,p)
    return LHS<=RHS
```

```python
#x0=(1.2,1.2)



x=np.array([1.2,1.2])
dx=np.array([np.inf,np.inf])
tol=1e-3
max_steps=100
rho=0.75
i=0


while np.linalg.norm(dx)>tol and i<max_steps:
    dx= Df5(x[0],x[1])
    
    a=1
    j=0
    
    while not WolfeI(a,f5,x,-dx):
        a*=rho
        j+=1
        
    xnew=x-a*dx
    
    x=xnew
    i+=1
    
    print(f'In iteration {i}, alpha={a} after {j} backtracks, and newx={x}')
    
print(f'After {i} iterations, approximate minimum is {f5(x[0],x[1])} at {x}')

```

For a quadratic function $f(\vec{x})=\frac{1}{2}\vec{x}^{T}Q\vec{x}-\vec{b}^{T}\vec{x}$, $Q$ is symmetric and positive-definite.We know $\nabla f_{k}=Q\vec{x}_{k}-\vec{b}$. Let $\vec{x}_{0}$ be the initial point. Since $\vec{x}^{*}$ is the minimizer, this satisfies $Q\vec{x}^{*}-\vec{b}=0$.

Consider 1 iteration with step length  $\alpha$:

$$\vec{x}_{1}=\vec{x}_{0}-\alpha \nabla f_{0}=\vec{x}_{0}-\alpha(Q\vec{x}_{0}-\vec{b})$$
or since $\vec{x}_{1}=\vec{x}^{*}$ (converges in 1 iteration).

$$\frac{\vec{x}_{0}-\vec{x}^{*}}{\alpha}=Q\vec{x}_{0}-\vec{b}=Q\vec{x}_{0}-Q\vec{x}^{*}=Q(\vec{x}_{0}-\vec{x}^{*})$$

Clearly this shows the iteration converges with 1 iteration if  $\vec{x}_{0}-\vec{x}^{*}$ is an eigenvector or is proportional to an eigenvector of $Q$ where the proportionality constant is $\alpha$ which becomes the step length.This is the case, granted, we read the proof backwards starting from the second equation at the far right. It is because the actual proof here ,if we start at with the first equation, shows that if the algorithm converges in 1 iteration, then $\vec{x}_{0}-\vec{x}^{*}$ is an eigenvector of $Q$.

```python
# Part b

f=lambda x,y:(x-1)**2+(2*y-1)**2
Df=lambda x,y: np.array([2*(x-1),4*(2*y-1)])

x=np.array([4,1])
dx=np.array([np.inf,np.inf])
path=[x]
tol=10**(-9)
alpha=13/125
max_steps=100
i=0

while np.linalg.norm(dx)>tol and i< max_steps:
    dx=Df(x[0],x[1])
    xnew=x-alpha*dx
    path.append(xnew)
    
    x=xnew
    i += 1
    
    
path=np.array(path) # sequence of points
print(path)
dif=np.diff(path,axis=0) # errors
print(dif)
err=np.linalg.norm(dif,axis=1)# 2-norm of the error terms
print(err)
S=err[i-1]/err[i-2] # This is the value of the ratio
print(S)

print(f'After {i} iterations, approximate minimum is {f(x[0],x[1])} at {x}')
```
