## 1)
 The given function is $$ f(x_{1},x_{2})=100(x_{2}-x_{1}^{2})^{2}+(1-x_{1})^{2}$$. We can compute the gradient as $\langle \frac{\partial f}{\partial x_{1}},\frac{\partial f}{\partial x_{2}} \rangle=\langle -400x_{1}(x_{2}-x_{1}^{2})-2(1-x_{1}),200(x_{2}-x_{1}^{2})\rangle$.
 
 The hessian matrix for this function is given by $$\begin{pmatrix} 1200x_{1}^{2}-400x_{2} +2 &-400x_{1} \\ -400x_{1} & 200 
 \end{pmatrix}$$

To help identify the minimizer, we must use the condition that $\nabla f=0$. Setting the second component equal to $0$, we see that $200(x_{2}-x_{1}^{2})=0$ or $x_{2}=x_{1}^{2}$. Plugging in this into the condition on the first component( $-400x_{1}(x_{2}-x_{1}^{2})-2(1-x_{1})=0$, we see that $-2(1-x_{1})=0$ or $x_{1}=1$. This implies that $x_{2}=1$. If we evlaute the hessian matrix given above at the point $(x_{1},x_{2})=(1,1)$, the matrix becomes :

$$\begin{pmatrix} 802 & -400 \\ -400 & 200 
 \end{pmatrix}$$

The eigenvalues $\lambda$ are given by $(800-\lambda)(200-\lambda)-160000=0$ The zeros are calculated below. We see that they are both positive and hence is positive definite. The point is thus the only minimizer.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

f=lambda x,y: 100*(y-x**2)**2+(1-x)**2
X=np.linspace(-3,3,100)
Y=np.linspace(-3,3,100)
Xmesh,Ymesh=np.meshgrid(X,Y)

Z=f(Xmesh,Ymesh)

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d', elev=20, azim=-60)
ax.plot_surface(Xmesh, Ymesh, Z, cmap='jet', alpha=0.5, vmax=250)
ax.set(xlabel='x', ylabel='y', zlabel='z', xlim=[-3,3], ylim=[-3,3], zlim=[-10000,10000])
plt.show()


plt.figure(figsize=(6,6))
CS=plt.contour(Xmesh,Ymesh,Z,5,cmap='jet',vmax=250)
plt.clabel(CS,inline_spacing=0,fmt='%d')
plt.axis([-3,3,-3,3])
plt.xlabel('x')
plt.ylabel('y')
plt.show()


```

# 2)

A function $f(x)$ has a stationary point at $x_{0}$ if $\nabla f(x_{0})=0$. Consider the function 
$f(x_{1},x_{2})=8x_{1}+12x_{2}+x_{1}^{2}-2x_{2}^{2}$. Cleary, $\nabla f(x)=\langle 8+2x_{1},12-4x_{2} \rangle$. This gradient is $0$ at $x=(-4,3)$. The Hessian matrix is

$$\begin{pmatrix} 2 & -4 \\ 0 & -4 
 \end{pmatrix}$$
 
 It is upper triangular so we can tell its eigenvalues are $2$ and $-4$. It is neither positive definite nor negative definite so it is indefinite. It is not a maximum or a minimum. The point must be a saddle point. 
 

```python
import numpy as np
import matplotlib.pyplot as plt

f1=lambda x,y: 8*x+12*y+x**2-2*y**2
X1=np.linspace(-20,20,100)
Y1=np.linspace(-20,20,100)
Xmesh1,Ymesh1=np.meshgrid(X1,Y1)

Z1=f1(Xmesh1,Ymesh1)

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection='3d', elev=20, azim=-60)
ax.plot_surface(Xmesh1, Ymesh1, Z1, cmap='jet', alpha=0.5, vmax=250)
ax.set(xlabel='x', ylabel='y', zlabel='z', xlim=[-20,20], ylim=[-20,20], zlim=[-400,1040])
plt.show()


plt.figure(figsize=(6,6))
CS=plt.contour(Xmesh1,Ymesh1,Z1,20,cmap='jet',vmax=250)
plt.clabel(CS,inline_spacing=0,fmt='%d')
plt.axis([-20,20,-20,20])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

We consider the function $f(x_{1},x_{2})=(x_{1}+x_{2}^{2})^{2}$. First we show that at $x=(1,0)$, $p=(-1,1)$ is a descent direction. First, we note the gradient is $\nabla f=\langle 2(x_{1}+x_{2}^{2}),4x_{2}(x_{1}+x_{2}^{2}) \rangle$. At $x=(1,0)$, this gradient is $(2,0)$. The condition for $p$ being a descent direction is that the dot product, $p * (2,0)<0$. Indeed, $(-1,1)*(2,0)=-2<0$ so this implies that $p$ is a descent direction at this specific point.

   To find the minimzers of the problem $min_{\alpha>0}f(x+\alpha p)$.Here $p=-\nabla f=\langle -2(x+y^{2}),-4y(x+y^{2})\rangle$.At the point $\langle x_{k},y_{k} \rangle$, the problem becomes $min_{\alpha >0}f(x+\alpha p)=f(\langle x_{k}-2\alpha(x_{k}+y_{k}^{2}),y_{k}(1-4\alpha)(x_{k}+y_{k}^{2}) \rangle)=min_{\alpha >0}[x_{k}-2\alpha(x_{k}+y_{k}^{2})+y_{k}^{2}(1-4\alpha)^{2}(x_{k}+y_{k}^{2})^{2}]^{2}$.
   
 Consider $$\phi(\alpha)= [x_{k}-2\alpha(x_{k}+y_{k}^{2})+y_{k}^{2}(1-4\alpha)^{2}(x_{k}+y_{k}^{2})^{2}]^{2}$$

To find the minimum of this one dimensional function in $\alpha$, we find the critical points of this function. The derivative with respect to $\alpha$ is set to $0$. This is 

$$32(1-4\alpha)y_{k}^{2}(x_{k}+y_{k}^{2})^{3}[x_{k}-2\alpha(x_{k}+y_{k}^{2})+(1-4\alpha)^{2}y_{k}^{2}(x_{k}+y_{k}^{2})^{2}]=0$$

This is satisfied by $\alpha=1/4$. We know that

$$[x_{k}-2\alpha(x_{k}+y_{k}^{2})+(1-4\alpha)^{2}y_{k}^{2}]=0$$

or rearranging this becomes

$$16\alpha^{2}y_{k}^{2}(x_{k}+y_{k}^{2})^{2}+\alpha(2(x_{k}+y_{k}^{2})-8y_{k}^{2}(x_{k}+y_{k}^{2})^{2})+(1-x_{k})=0$$

This has two solutions:

$$\alpha_{1}=\frac{-2(x_{k}+y_{k}^{2})+8y_{k}^{2}(x_{k}+y_{k}^{2})^{2}+\sqrt{[2(x_{k}+y_{k}^{2}-8y_{k}^{2}(x_{k}+y_{k}^{2})^{2}]^{2}-64y_{k}^{2}(x_{k}+y_{k}^{2})^{2}(1-x_{k})}}{32y_{k}^{2}(x_{k}+y_{k}^{2})^{2}}$$

$$\alpha_{2}=\frac{-2(x_{k}+y_{k}^{2})+8y_{k}^{2}(x_{k}+y_{k}^{2})^{2}-\sqrt{[2(x_{k}+y_{k}^{2})-8y_{k}^{2}(x_{k}+y_{k}^{2})^{2}]^{2}-64y_{k}^{2}(x_{k}+y_{k}^{2})^{2}(1-x_{k})}}{32y_{k}^{2}(x_{k}+y_{k}^{2})^{2}}$$

For the specific descent direction at $p=(-1,1)$. An analogous process is carried out. In particular, the minimization is 

$$min_{\alpha>0}(x_{k}-\alpha+(y_{k}+\alpha)^{2})^{2}$$

The derivavtive with respect to $\alpha$ and setting equal to zero implies the follwoing must be zero

$$4(y_{k}+\alpha)(x_{k}-\alpha+(y_{k}+\alpha)^{2})=0$$

or 

$$\alpha=-y_{k}$$

Rearrainging the quadratic:

$$ \alpha^{2}+\alpha(2y_{k}-1)+(x_{k}+y_{k}^{2})=0 $$

The roots are

$$ \alpha=\frac{1-2y_{k}\pm \sqrt{(2y_{k}-1)^{2}-4(x_{k}+y_{k}^{2})}}{2}$$

It is found that starting at $x_{0}=(1.2,1.2)$ with step length $\alpha=10^{-3}$, the number of iterations needed so that $|p|=|-\nabla f|<10^{-9}$ is $46566$. If the step size is made larger, $\alpha=10^{-2}$, an overflow error is encountered after a small number of iterations idicating that the function has attained fantastically large values(~$10^{136}$) and so the method does not converge to the minimum($0$). The second round of runs has the initial point at $x_{0}=(-1.2,1)$. The number of iterations that satisfy the same conditions above is $49371$. If we choose an even bigger step length at $\alpha=10^{-1}$, the method diverges at a faster rate and an overflow error is encountered.

```python
plt.figure(figsize=(6, 6))
CS = plt.contour(Xmesh, Ymesh, Z, np.arange(2,18,2)**2, cmap='jet')
plt.clabel(CS,inline_spacing=0,fmt='%d')
plt.axis([-3,3,-3,3])
plt.xlabel('x')
plt.ylabel('y')


f5=lambda x: 100*(x[1]-x[0]**2)**2+(1-x[0])**2

Df5=lambda x: np.array([-400*x[0]*(x[1]-x[0]**2)-2*(1-x[0]),200*(x[1]-x[0]**2)])

x=np.array([1.2,1.2])
dx=np.array([np.inf,np.inf])
tol=10**(-9)
alpha=10**(-3)
max_steps=50000
i=0

while np.linalg.norm(dx)>tol and i< max_steps:
    dx=Df5(x)
    xnew=x-alpha*dx
    plt.arrow(x[0],x[1],-alpha*dx[0],-alpha*dx[1],head_width=.25,head_length=.25,color= 'b')
    x=xnew
    i += 1
    
print('After '+str(i)+' iterations, approximate minimum is '+str(f5(x))+' at ('+str(x[0])+','+str(x[1])+')')    
plt.show()
```
