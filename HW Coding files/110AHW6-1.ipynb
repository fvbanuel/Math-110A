It is found that the greatest value corresponds to the line of slope $-5$ with greatest y-intercept within the feasible region shaded in grey. This line is at the edge of the region and corresponds to the line $5x+y=20$ so that the maximum is of the objective is $20$.

The objective is 

$$\max 5x_{1}+ x_{2}$$

subject to

$$ x_{1}+x_{2}\leq 5$$

$$ 2x_{1}+0.5x_{2}\leq 8$$

```python
plt.figure(figsize=(5,5),dpi=150)
x = np.linspace(-1,6,200)
y1 = 5 - x        
y2 = 16-4*x   
y3=20-5*x 



plt.plot(x,y1,x,y2,x,y3)
plt.plot(x,x*0,'k',x*0,y2,'k')
upper = np.min(np.array([y1,y2]),axis=0)
plt.fill_between(x, 0*x, upper, where=(upper >= 0)*(x>=0), color='grey', alpha=0.5) # shades feasible region
plt.legend(['x+y=5','2x+0.5y=8','Objec. Maximizer'])
plt.show()
```

The above is the same as 

$$\min -5x_{1}-x_{2}$$

subject to

$$ x_{1}+x_{2}\leq 5$$

$$ 2x_{1}+0.5x_{2}\leq 8$$

$$ x_{1}\geq 0$$
$$ x_{2} \geq 0$$

Now introduce the slack variables $x_{3},x_{4}$ making the problem as

$$\min -5x_{1}-x_{2}$$

subject to

$$ x_{1}+x_{2}+x_{3}= 5$$

$$ 2x_{1}+0.5x_{2}+x_{4}= 8$$

$$ x_{1}\geq 0$$
$$ x_{2} \geq 0$$
$$ x_{3} \geq 0$$
$$ x_{4} \geq 0$$.


The origin is a feasible solution so this will be the initial solution:$(0,0,5,8)$.Now write basis variables in terms of non-basis variables.

$$x_{3}=5-x_{1}-x_{2}\geq 0$$
$$x_{4}=8-2x_{1}-0.5x_{2}\geq 0$$

Looking at the objective, we let $x_{1}$ enter the basis. Now check the constraints by letting $x_{2}=0$. This is gives

$$x_{3}=5-x_{1}\geq0    \Rightarrow  x_{1}\leq 5$$
$$x_{4}=8-2x_{1}\geq 0   \Rightarrow x_{1}\leq 4$$

So we should let $x_{1}=4$ and thus $x_{4}=0$, so that $x_{4}$ leaves the basis. Now rewrite the objective in terms of the new non-basis variables and the basis variables in terms  of non-basis varibales.

$$x_{1}=4-\frac{1}{2}x_{4}-\frac{1}{4}x_{2}$$
$$x_{3}=5-(4-\frac{1}{2}x_{4}-\frac{1}{4}x_{2})-x_{2}=1+\frac{1}{2}x_{4}-\frac{3}{4}x_{2}$$

The objective is

$$-5x_{1}-x_{2}=-5(4-\frac{1}{2}x_{4}-\frac{1}{4}x_{2})-x_{2}=-20+\frac{5}{2}x_{4}+\frac{1}{4}x_{2}$$

No further iteration is needed as the coefficients of the non-basis variables are all positive. The minimum is thus $-20$ or a maximum of $20$. The point corresponding to this is$(4,0,5,0)$or $(4,0)$ in the $x_{1}x_{2}$ plane. This was verified graphically above.
