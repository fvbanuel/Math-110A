### Example: Plotting a function of two variables

Consider $f(x,y) = x^2 + 2y^2 + 4x + y + 6$. Below, we graph this function in 3D using the `mpl_toolkits.mplot3d` module provided by `matplotlib`.

```python
import numpy as np  # arrays
import matplotlib.pyplot as plt  # 2d plotting
from mpl_toolkits.mplot3d import axes3d  # 3d plotting

f = lambda x, y: x**2 + 2*y**2 + 4*x + y + 6
f(1,-1)
```

```python
X = np.linspace(-10,10,300)  # 300 evenly spaced points on x-axis [-10,10]
Y = np.linspace(-10,10,300)  # 300 evenly spaced points on y-axis [-10,10]
print(X)
```

```python
# Mathematical operations in definition of f act componentwise on NumPy arrays
X**2
# 2*Y
# X+Y
```

```python
f(X,Y)  # evaluate at (X[0],Y[0]), then (X[1],Y[1])), and so on
```

```python
Xmesh, Ymesh = np.meshgrid(X,Y)  # creates 300x300 grid of points defined by X and Y above
print(Xmesh.shape)
print(Xmesh[:5,:5])
print(Ymesh[:5,:5])
```

```python
Z = f(Xmesh, Ymesh)  # evaluate f at each of the 300x300 points
print(Z[:5,:5])
```

```python
fig = plt.figure(figsize=(10, 6))
# feel free to play with the arguments below to get a feel for 3D plotting:
#     elev is angle above xy-plane, azim is angle around z-axis
#     Ex: elev=0, azim=0 shows yz-plane with positive x-axis out of screen 
ax = plt.axes(projection='3d', elev=30, azim=-60)
ax.plot_surface(Xmesh, Ymesh, Z, cmap='jet', alpha=0.5, vmax=250)
ax.set(xlabel='x', ylabel='y', zlabel='z', xlim=[-10,10], ylim=[-10,10], zlim=[0,256])
plt.show()
```

```python
plt.figure(figsize=(6, 6))
CS = plt.contour(X, Y, Z, np.arange(2,18,2)**2, cmap='jet',vmax=250) # fourth argument is where to draw contour lines
# label the contours along the lines; without inline_spacing, too much white space
plt.clabel(CS,inline_spacing=0,fmt='%d')
plt.axis([-10,10,-10,10])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Above, we've used the `jet` colormap. See [here](https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html) for many other options.

## Iterative optimization algorithms

We stated last time that it is very easy conceptually to state a necessary and sufficient condition for a point to be a local minimum of a function. We must have $\nabla f(\mathbf{x})=\mathbf{0}$ and $\nabla^2f(\mathbf{x})>0$. We even showed an analytic example where solving the system $\nabla f=\mathbf{0}$ was relatively straightforward. In many if not most applications, though, solving this system of in general nonlinear equations is very difficult. Most of the optimization algorithms we discuss in this class will be **iterative optimization algorithms**, where we make some initial "guess" of solution $\mathbf{x}_0$, then iteratively update our guess to create a sequence $\mathbf{x}_1$, $\mathbf{x}_2$, $\ldots$, $\mathbf{x}_k$, $\ldots$ which hopefully better and better approximates a minimum value of the function $f$. That is, we hope to satisfy $f(\mathbf{x}_1)\ge f(\mathbf{x}_2)\ge\cdots\ge f(\mathbf{x}_k)\ge\cdots$ so that we're constantly **decreasing** the value of the function with each iteration, i.e. **descending**.

There are two main categories of iterative optimization algorithms:

#### 1) Line search algorithms

These are the most conceptually straightforward algorithms, *gradient descent* being the simplest and therefore most commonly used. The main idea is to choose some direction $\mathbf{p}_k$ in which to 
"step" at each iteration and set

$$ \mathbf{x}_{k+1}=\mathbf{x}_k+\alpha_k\mathbf{p}_k $$

for some scalar value $\alpha_k$ called the "step length", determined at each iteration. If the direction $\mathbf{p}_k$ and step length $\alpha_k$ are chosen wisely, this iteration can be shown analytically to converge under certain conditions. The choice of $\mathbf{p}_k$ is often (e.g. in gradient descent) $\mathbf{p}_k=-\nabla f(\mathbf{x}_k)$ since calculus tells us the gradient is the direction of steepest *ascent* and thus it makes sense to move in the *opposite* direction to *descend*. Indeed, any direction $\mathbf{p}_k$ with $\mathbf{p}_k\cdot\nabla f(\mathbf{x}_k)<0$ is guaranteed to decrease the value of the function, at least in the limit $\alpha_k\to0$. There are some more complicated algorithms which try to more cleverly choose $\mathbf{p}_k$ to attempt to overcome issues with pure gradient descent, particularly for very non-convex functions, which we may discuss later.

A typically more difficult problem than choosing the best step *direction* is choosing the best step *length* $\alpha_k$. Given some choice of $\mathbf{p}_k$, if we wanted to be as rigorous as possible, we could set
$$ \alpha_k = \min_{\alpha>0} f(\mathbf{x}_k+\alpha \mathbf{p}_k) $$

i.e., look infinitely far along the line in the direction of $\mathbf{p}_k$ and find the smallest value of $f$ in that direction. To find the value of $\alpha$ that minimizes the RHS, we could think of it as a (one-dimensional!) function of $\alpha$, say, $\phi(\alpha)=f(\mathbf{x}+\alpha \mathbf{p}_k)$ and find where $\phi'(\alpha)=0$. This at the very least turns our high-dimensional optimization problem into a 1-D problem, but even this can be difficult to solve in general. Instead of always looking for the *best* $\alpha_k$, we often settle for a "*good enough*" $\alpha_k$, indeed usually just setting $\alpha_k$ equal to the same constant for all $k$. The simplest form of this so called **fixed step** descent is classic **gradient descent**,
$$ \mathbf{x}_{k+1}=\mathbf{x}_k-\alpha\nabla f(\mathbf{x}_k),\qquad \alpha\in\mathbb{R}$$

#### 2) Trust region algorithms

We will begin the quarter talking about line search algorithms, so we only say a little about trust region methods here. The basic idea behind these methods is the following: Instead of first choosing a direction and finding the optimal distance to step in that direction, first choose a step length, then find the optimal direction in a neighborhood around the current iterate. That is, solve the optimization problem

$$ \mathbf{p}_k = \min_{\|\mathbf{p}\|<\alpha_k} f(\mathbf{x}_k+\mathbf{p}) $$

Instead of simplifying the high-dimensional problem by looking only in one direction from a given iterate, trust region methods usually simply the problem by making the form of $f$ "nicer" in a certain region through approximations. That is, replace $f$ with e.g. its Taylor series expansion $f\approx m_k$ at each step, and take advantage of the nice polynomial form to determine a general solution to the above.

### Example: Gradient descent

Let us look again at the function $f(x,y)=x^2 + 2y^2 + 4x + y + 6$ plotted above and perform basic gradient descent ($\mathbf{p}_k=-\nabla f(\mathbf{x}_k)$) with $\alpha_k=0.1$ for all $k$, starting from an initial guess of $\mathbf{x}_0=(3,-2)$ below:

```python
# plot the figure first
plt.figure(figsize=(6, 6))
CS = plt.contour(X, Y, Z, np.arange(2,18,2)**2, cmap='jet')
plt.clabel(CS,inline_spacing=0,fmt='%d')
plt.axis([-10,10,-10,10])
plt.xlabel('x')
plt.ylabel('y')

f = lambda x: x[0]**2 + 2*x[1]**2 + 4*x[0] + x[1] + 6
# define derivatives of f to make the gradient
Df = lambda x: np.array([2*x[0]+4, 4*x[1]+1])

x = np.array([3,-2])  # initial point
dx = np.array([np.inf,np.inf]) # initial large gradient so while loop runs
alpha = .1            # step size
tol = 1e-3            # stop when gradient is smaller than this amount
max_steps = 100       # Maximum number of steps to run the iteration
i=0                   # iteration count
while np.linalg.norm(dx)>tol and i<max_steps:
    dx = Df(x)
    # new value of x
    xnew = x - alpha*dx
    # add arrow to plot
    plt.arrow(x[0],x[1],-alpha*dx[0],-alpha*dx[1],head_width=.25,head_length=.25,color='b')
    # update old value
    x = xnew
    # update iteration count
    i += 1

print('After '+str(i)+' iterations, approximate minimum is '+str(f(x))+' at ('+str(x[0])+','+str(x[1])+')')    
plt.show()
```
